```python
import numpy as np
import matplotlib.pyplot as plt

def plot_akurasi_per_fold(
    akurasi_fold,
    nama_model="Model",
    y_min=0.800,
    y_max=1.000,
    figsize=(9, 5),
):
    akurasi_fold = np.array(akurasi_fold, dtype=float)
    fold_ke = np.arange(1, len(akurasi_fold) + 1)

    rata_rata = akurasi_fold.mean()
    nilai_tertinggi = akurasi_fold.max()
    fold_tertinggi = fold_ke[akurasi_fold.argmax()]
    nilai_terendah = akurasi_fold.min()
    fold_terendah = fold_ke[akurasi_fold.argmin()]

    plt.figure(figsize=figsize)

    # Line chart
    plt.plot(
        fold_ke, akurasi_fold,
        marker="o", linewidth=2,
        label="Akurasi per Fold",
    )

    # Garis rata-rata (putus-putus)
    plt.axhline(
        rata_rata,
        linestyle="--", linewidth=1.8, color="gray",
        label=f"Rata-rata: {rata_rata:.4f}",
    )

    # Garis nilai tertinggi (putus-putus)
    plt.axhline(
        nilai_tertinggi,
        linestyle="--", linewidth=1.8, color="red",
        label=f"Fold tertinggi (Fold {fold_tertinggi}): {nilai_tertinggi:.4f}",
    )

    # Titik fold tertinggi
    plt.scatter([fold_tertinggi], [nilai_tertinggi], color="red", zorder=5)

    # Garis nilai terendah (putus-putus)
    plt.axhline(
        nilai_terendah,
        linestyle="--", linewidth=1.8, color="blue",
        label=f"Fold terendah (Fold {fold_terendah}): {nilai_terendah:.4f}",
    )

    # Titik fold terendah
    plt.scatter([fold_terendah], [nilai_terendah], color="blue", zorder=5)

    plt.title(f"Akurasi tiap fold {nama_model}")
    plt.xlabel("Fold ke-")
    plt.ylabel("Akurasi")

    plt.xticks(fold_ke)
    plt.grid(True, linestyle=":", alpha=0.6)
    plt.ylim(y_min, y_max)

    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_akurasi_per_fold([0.9426, 0.9342, 0.9616, 0.9370, 0.9644], nama_model="KNN", y_min=0.8000, y_max=1.0000)
```
